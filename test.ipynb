```python
from ultralytics import YOLO
import cv2
from PIL import Image


# Load a pretrained YOLO11n model
model = YOLO("app/Models/yolov9_objectLocalizer.pt")

# Define path to video file
source = "Videos/Month_1/5.22.21-7-21-Testing.MOV"
out = "out_object.mp4"
# Run inference on the source
results = model(source, stream=True)  # generator of Results objects
# Setup videowriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # You can change 'XVID' to other codecs (e.g., 'mp4v' for .mp4)
fps = 30  # Frames per second
imgs=[]
r = next(results)
im_bgr = r.plot() 
height, width, layers = im_bgr.shape

video_writer = cv2.VideoWriter(out, fourcc, fps, (width, height))

for i, r in enumerate(results):
    # Plot results image
    im_bgr = r.plot()  # BGR-order numpy array
    
    
    video_writer.write(im_bgr)
video_writer.release()
```
